## üìä Overview
### This project builds a classification model to predict the species of Iris flowers based on four key measurements: sepal length, sepal width, petal length, and petal width. We use a Random Forest Classifier to perform the classification and evaluate model performance.


## üìÅ Dataset

### - Sepal Length

### - Petal Width

## üß† Code
# ```python

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
````

````python
# Load dataset
df = pd.read_csv(r"C:\Users\hp\Downloads\archive (1)\IRIS.csv")

# Preview data
print("First 5 rows of the dataset:")
print(df.head())
````

### Data visualization

````python
sns.pairplot(df, hue="species")
plt.suptitle("Pairplot of Iris Features", y=1.02)
plt.show()
````

````python
# Encode target labels
df["species"] = df["species"].astype("category").cat.codes
````

````python
# Features and labels
X = df.drop("species", axis=1)
y = df["species"]
````

````python
# Normalize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
````

````python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Train model
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
````

````python
# Predict
y_pred = model.predict(X_test)
````

````python
# Evaluation
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
````

````python
# Feature Importance
importances = model.feature_importances_
feature_names = df.columns[:-1]
sns.barplot(x=importances, y=feature_names)
plt.title("Feature Importance")
plt.show()
````
